### Questions ouvertes

```python
# Combien de médailles la France a remporté : en tout, en Or, en argent et en Bronze (depuis le début des JO) ?

# filtre sur la France
france_medals = df_medals[df_medals['country_name'] == 'France']

# nombre total de médailles
total_medals = france_medals.shape[0]

#nombre de médailles d'or, d'argent et de bronze
gold_medals = france_medals[france_medals['medal_type'] == 'GOLD'].shape[0]
silver_medals = france_medals[france_medals['medal_type'] == 'SILVER'].shape[0]
bronze_medals = france_medals[france_medals['medal_type'] == 'BRONZE'].shape[0]

print(f"Total des médailles remportées par la France : {total_medals}")
print(f"Total des médailles d'or : {gold_medals}")
print(f"Total des médailles d'argent : {silver_medals}")
print(f"Total des médailles de bronze : {bronze_medals}")
```

```python
# Lors de quelle JO la France a eu le plus (le moins) de succès ?

#nombre de médailles pour chaque édition
medals_per_game = france_medals.groupby('game_slug').size().reset_index(name='medal_count')

max_medals = medals_per_game[medals_per_game['medal_count'] == medals_per_game['medal_count'].max()]

min_medals = medals_per_game[medals_per_game['medal_count'] == medals_per_game['medal_count'].min()]

print("Édition des JO avec le plus de succès pour la France:")
print(max_medals)
print("Édition des JO avec le moins de succès pour la France:")
print(min_medals)
```

```python
# nombre de médaille obtenues par la France pour chaque JO
medals_per_game_sorted = medals_per_game.sort_values(by='medal_count')
medals_per_game_sorted
```

```python
# Peut-on considérer que la France est la grande spécialiste de certaine(s) discipline(s) sportive(s),
# comparativement aux autres nations ? Laquelle / Lesquelles ?

# nombre de médailles par discipline pour la France
medals_per_discipline = france_medals.groupby('discipline_title').size().reset_index(name='medal_count')

excellent_disciplines = medals_per_discipline[medals_per_discipline['medal_count'] >= 50]

print("Meilleur Disciplines de la France :")
print(excellent_disciplines)
```

```python
# Quelles sont les sports les plus dominants dans les JO au fil des ans ?

#nombre total de médailles par sport
medals_per_sport = df_medals.groupby('discipline_title').size().reset_index(name='medal_count')

#trier les sports par le nombre total de médailles
most_dominant_sports = medals_per_sport.sort_values(by='medal_count', ascending=False)

print("Sports les plus dominants dans les Jeux Olympiques :")
most_dominant_sports.head(10)
```

```python
# top 25 des pays par nombre total de médailles
top_25_countries = df_medals['country_name'].value_counts().head(25).index
top_25_medals = df_medals[df_medals['country_name'].isin(top_25_countries)]

#grouper les données par pays et type de médaille
top_25_medals_grouped = top_25_medals.groupby(['country_name', 'medal_type']).size().unstack(fill_value=0).reset_index()

print(top_25_medals_grouped)
```

```python
# 2. La France est le 2è pays qui a organisé le plus de JO après les USA (8 JO) ?

# Compter le nombre de fois que chaque pays a organisé les JO
hosts_count = df_hosts['game_location'].value_counts()

# Trouver le rang de la France et comparer au nombre total pour les USA
france_rank = hosts_count.rank(method='max', ascending=False)['France']
usa_organized = hosts_count['United States']

# Afficher les résultats
print(f"La France est classée {int(france_rank)}ème en nombre de JO organisés.")
print(f"Les États-Unis ont organisé {usa_organized} JO.")
```

```python
# 3. Les JO d’hiver sont nés à Chamonix en 1924 ?

# Filtrer les Jeux Olympiques d'hiver
winter_olympics = df_hosts[df_hosts['game_season'] == 'Winter']

# Trier par année pour trouver le premier Jeux d'hiver organisé
first_winter_olympics = winter_olympics.sort_values(by='game_year').iloc[0]

# Construire le message de réponse
location = first_winter_olympics['game_location']
year = first_winter_olympics['game_year']
game_name = first_winter_olympics['game_name']

# Afficher le résultat
print(f"Les premiers JO d'Hiver ont eu lieu à {game_name} en {year}")
```

```python
# 4. JO de Paris, en 1900 : les femmes peuvent participer aux JO ?

# Filtrer les médailles remportées lors des Jeux de Paris 1900
paris_1900_medals = df_medals[df_medals['game_slug'] == 'paris-1900']

# Vérifier la présence de compétitions féminines
women_events = paris_1900_medals[paris_1900_medals['event_gender'] == 'Women']

# Afficher les résultats
if not women_events.empty:
    print(f"Des compétitions féminines ont eu lieu lors des JO de Paris 1900.")
    print(f"Événements: {women_events['event_title'].unique()}")
else:
    print("Aucune compétition féminine n'a été enregistrée lors des JO de Paris 1900.")
```

```python
# 6. De 1924 à 1992, les JO d’hiver et d’été avaient lieu au cours de la même année. Désormais, ils sont organisés selon des cycles distincts avec une alternance de 2 ans ?

# Convertir les années en entiers
df_hosts['game_year'] = df_hosts['game_year'].astype(int)

# Filtrer les données pour les Jeux entre 1924 et 1992
games_1924_1992 = df_hosts[(df_hosts['game_year'] >= 1924) & (df_hosts['game_year'] <= 1992)]

# Vérifier si les Jeux d'hiver et d'été avaient lieu la même année
same_year_games = games_1924_1992.pivot_table(index='game_year', columns='game_season', aggfunc='size', fill_value=0)
same_year_games = same_year_games[(same_year_games['Summer'] > 0) & (same_year_games['Winter'] > 0)]

# Afficher les résultats
if same_year_games.empty:
    print("Entre 1924 et 1992, les JO d'hiver et d'été n'ont pas eu lieu la même année.")
else:
    print(same_year_games)
    print("Entre 1924 et 1992, les JO d'hiver et d'été ont bien eu lieu la même année.")


# Vérifier s'ils sont désormais organisés selon des cycles distincts avec une alternance de 2 ans après 1992
games_after_1992 = df_hosts[df_hosts['game_year'] > 1992]
games_after_1992 = games_after_1992.pivot_table(index='game_year', columns='game_season', aggfunc='size', fill_value=0)

# Afficher les résultats
if games_after_1992.empty:
    print("Après 1992, les JO d'hiver et d'été n'ont pas eu lieu la même année.")
else:
    print(games_after_1992)
    print("Après 1992, les JO d'hiver et d'été sont organisés selon des cycles distincts avec une alternance de 2 ans.")
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
```

```python
# normalisation des données
scaler = StandardScaler()
X_all_train = scaler.fit_transform(X_all_train)
X_all_test = scaler.transform(X_all_test)
```

```python
# Entraînement du modèle MLP pour le totals des médailles 
model_mlp_total = Sequential([
    Dense(128, activation='relu', input_shape=(X_all_train.shape[1],)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='linear')
])
model_mlp_total.compile(optimizer='adam', loss='mse', metrics=['mae'])
model_mlp_total.fit(X_all_train, y_all_total_train, epochs=10, batch_size=32, validation_split=0.2)

# Entraînement du modèle MLP pour les médailles d'or
model_mlp_gold = Sequential([
    Dense(128, activation='relu', input_shape=(X_all_train.shape[1],)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='linear')
])
model_mlp_gold.compile(optimizer='adam', loss='mse', metrics=['mae'])
model_mlp_gold.fit(X_all_train, y_all_gold_train, epochs=10, batch_size=32, validation_split=0.2)

# Entraînement du modèle MLP pour les médailles d'argent
model_mlp_silver = Sequential([
    Dense(128, activation='relu', input_shape=(X_all_train.shape[1],)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='linear')
])
model_mlp_silver.compile(optimizer='adam', loss='mse', metrics=['mae'])
model_mlp_silver.fit(X_all_train, y_all_silver_train, epochs=10, batch_size=32, validation_split=0.2)

# Entraînement du modèle MLP pour les médailles de bronze
model_mlp_bronze = Sequential([
    Dense(128, activation='relu', input_shape=(X_all_train.shape[1],)),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='linear')
])
model_mlp_bronze.compile(optimizer='adam', loss='mse', metrics=['mae'])
model_mlp_bronze.fit(X_all_train, y_all_bronze_train, epochs=10, batch_size=32, validation_split=0.2)
```

### Réseau de Neurones Convolutionnels - CNN

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten
```

```python
#reshape des données pour le CNN
X_all_train_cnn = X_all_train.reshape(X_all_train.shape[0], X_all_train.shape[1], 1)
X_all_test_cnn = X_all_test.reshape(X_all_test.shape[0], X_all_test.shape[1], 1)

# Entraînement du modèle CNN pour le total des médailles
model_cnn_total = Sequential([
    Conv1D(32, 3, activation='relu', input_shape=(X_all_train_cnn.shape[1], 1)),
    MaxPooling1D(2),
    Dropout(0.5),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='linear')
])
model_cnn_total.compile(optimizer='adam', loss='mse', metrics=['mae'])
model_cnn_total.fit(X_all_train_cnn, y_all_total_train, epochs=10, batch_size=32, validation_split=0.2)

# Entraînement du modèle CNN pour les médailles d'or
model_cnn_gold = Sequential([
    Conv1D(32, 3, activation='relu', input_shape=(X_all_train_cnn.shape[1], 1)),
    MaxPooling1D(2),
    Dropout(0.5),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='linear')
])
model_cnn_gold.compile(optimizer='adam', loss='mse', metrics=['mae'])
model_cnn_gold.fit(X_all_train_cnn, y_all_gold_train, epochs=10, batch_size=32, validation_split=0.2)

# Entraînement du modèle CNN pour les médailles d'argent
model_cnn_silver = Sequential([
    Conv1D(32, 3, activation='relu', input_shape=(X_all_train_cnn.shape[1], 1)),
    MaxPooling1D(2),
    Dropout(0.5),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='linear')
])
model_cnn_silver.compile(optimizer='adam', loss='mse', metrics=['mae'])
model_cnn_silver.fit(X_all_train_cnn, y_all_silver_train, epochs=10, batch_size=32, validation_split=0.2)

# Entraînement du modèle CNN pour les médailles de bronze
model_cnn_bronze = Sequential([
    Conv1D(32, 3, activation='relu', input_shape=(X_all_train_cnn.shape[1], 1)),
    MaxPooling1D(2),
    Dropout(0.5),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='linear')
])
model_cnn_bronze.compile(optimizer='adam', loss='mse', metrics=['mae'])
model_cnn_bronze.fit(X_all_train_cnn, y_all_bronze_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
predictions_cnn_total = model_cnn_total.predict(X_all_test)
predictions_cnn_gold = model_cnn_gold.predict(X_all_test)
predictions_cnn_silver = model_cnn_silver.predict(X_all_test)
predictions_cnn_bronze = model_cnn_bronze.predict(X_all_test)

print('Predicted labels total : ', np.round(predictions_cnn_total)[:10])
print('Actual labels total : ' , y_all_gold_test[:10])
print('Predicted labels gold : ', np.round(predictions_cnn_gold)[:10])
print('Actual labels gold: ' , y_all_total_test[:10])
print('Predicted labels silver : ', np.round(predictions_cnn_silver)[:10])
print('Actual labels silver : ' , y_all_silver_test[:10])
print('Predicted labels bronze : ', np.round(predictions_cnn_bronze)[:10])
print('Actual labels bronze : ' , y_all_bronze_test[:10])
```

```python
mse = mean_squared_error(y_all_total_test, predictions_cnn_total)
print("MSE total:", mse)
rmse = np.sqrt(mse)
print("RMSE total:", rmse)
r2 = r2_score(y_all_total_test, predictions_cnn_total)
print("R2 total:", r2)

mse = mean_squared_error(y_all_gold_test, predictions_cnn_gold)
print("\nMSE gold:", mse)
rmse = np.sqrt(mse)
print("RMSE gold:", rmse)
r2 = r2_score(y_all_gold_test, predictions_cnn_gold)
print("R2 gold:", r2)


mse = mean_squared_error(y_all_silver_test, predictions_cnn_silver)
print("\nMSE silver:", mse)
rmse = np.sqrt(mse)
print("RMSE silver:", rmse)
r2 = r2_score(y_all_gold_test, predictions_cnn_silver)
print("R2 silver:", r2)


mse = mean_squared_error(y_all_bronze_test, predictions_cnn_bronze)
print("\nMSE bronze:", mse)
rmse = np.sqrt(mse)
print("RMSE bronze:", rmse)
r2 = r2_score(y_all_gold_test, predictions_cnn_bronze)
print("R2 bronze:", r2)
```
