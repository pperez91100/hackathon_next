```python
import xml.etree.ElementTree as ET
import json
import pandas as pd
```

```python
#JSON
json_file_path = './Hackathon_Data/olympic_athletes.json'

# Ouvrir et lire le fichier JSON
with open(json_file_path, 'r', encoding='utf-8') as file:
    json_data = json.load(file)

print(json_data[:3])
```

```python
df_athlete = pd.DataFrame(json_data)
df_athlete
```

```python
#XML
xml_file_path = './Hackathon_Data/olympic_hosts.xml'

# Lire et parser le fichier XML
tree = ET.parse(xml_file_path)
root = tree.getroot()

#convertir les données en une structure de dictionnaire
data = []
for row in root.findall('row'):
    row_data = {subchild.tag: subchild.text for subchild in row}
    data.append(row_data)
```

```python
# xml dans un dataframe
df_hosts = pd.DataFrame(data)
df_hosts
```

```python
# Combien de médailles la France a remporté : en tout, en Or, en argent et en Bronze (depuis le début des JO) ?

# on filtre sur la France
france_medals = df_medals[df_medals['country_name'] == 'France']

# nombre total de médailles
total_medals = france_medals.shape[0]

#nombre de médailles d'or, d'argent et de bronze
gold_medals = france_medals[france_medals['medal_type'] == 'GOLD'].shape[0]
silver_medals = france_medals[france_medals['medal_type'] == 'SILVER'].shape[0]
bronze_medals = france_medals[france_medals['medal_type'] == 'BRONZE'].shape[0]

print(f"Total des médailles remportées par la France : {total_medals}")
print(f"Total des médailles d'or : {gold_medals}")
print(f"Total des médailles d'argent : {silver_medals}")
print(f"Total des médailles de bronze : {bronze_medals}")
```

```python
# Lors de quelle JO la France a eu le plus (le moins) de succès ?

# filtre sur la France
france_medals = df_medals[df_medals['country_name'] == 'France']

#nombre de médailles pour chaque édition
medals_per_game = france_medals.groupby('slug_game').size().reset_index(name='medal_count')

# le plus de médailles
max_medals = medals_per_game[medals_per_game['medal_count'] == medals_per_game['medal_count'].max()]

#le moins de médailles
min_medals = medals_per_game[medals_per_game['medal_count'] == medals_per_game['medal_count'].min()]

print("Édition des JO avec le plus de succès pour la France:")
print(max_medals)
print("Édition des JO avec le moins de succès pour la France:")
print(min_medals)
```

```python
# nombre de médaille obtenues par la France pour chaque JO
medals_per_game_sorted = medals_per_game.sort_values(by='medal_count')
medals_per_game_sorted
```

```python
# Peut-on considérer que la France est la grande spécialiste de certaine(s) discipline(s) sportive(s),
# comparativement aux autres nations ? Laquelle / Lesquelles ?

# nombre de médailles par discipline
medals_per_discipline = france_medals.groupby('discipline_title').size().reset_index(name='medal_count')

excellent_disciplines = medals_per_discipline[medals_per_discipline['medal_count'] >= 50]

print("Meilleur Disciplines de la France :")
print(excellent_disciplines)
```

```python
# Quelles sont les sports les plus dominants dans les JO au fil des ans ?

#nombre total de médailles par sport
medals_per_sport = df_medals.groupby('discipline_title').size().reset_index(name='medal_count')

#trier les sports par le nombre total de médailles
most_dominant_sports = medals_per_sport.sort_values(by='medal_count', ascending=False)

print("Sports les plus dominants dans les Jeux Olympiques :")
most_dominant_sports.head(10)
```

## Questions - Evènements marquants aux JO :

```python
# 1. La France a organisé 6 JO : 3 d’hiver et 3 d’été (en comptant celui de 2024) ?

# Filtrer les Jeux Olympiques organisés par la France en utilisant la colonne correcte 'game_location'
france_olympics = df_hosts[df_hosts['game_location'] == 'France']

# Séparer les Jeux d'été et d'hiver en utilisant la colonne 'game_season'
summer_olympics = france_olympics[france_olympics['game_season'] == 'Summer']
winter_olympics = france_olympics[france_olympics['game_season'] == 'Winter']

# Compter le nombre de Jeux d'été et d'hiver
summer_count = len(summer_olympics)
winter_count = len(winter_olympics)
total_count = summer_count + winter_count

print(f"Nombre de Jeux d'été organisés en France : {summer_count}")
print(f"Nombre de Jeux d'hiver organisés en France : {winter_count}")
print(f"Total des Jeux Olympiques organisés en France : {total_count}")
```

##### En ajoutant les JO 2024, on a donc bien 6 JO organisés par la France

```python
# 2. La France est le 2è pays qui a organisé le plus de JO après les USA (8 JO) ?

# Compter le nombre de fois que chaque pays a organisé les JO
hosts_count = df_hosts['game_location'].value_counts()

# Trouver le rang de la France et comparer au nombre total pour les USA
france_rank = hosts_count.rank(method='max', ascending=False)['France']
usa_organized = hosts_count['United States']

# Afficher les résultats
print(f"La France est classée {int(france_rank)}ème en nombre de JO organisés.")
print(f"Les États-Unis ont organisé {usa_organized} JO.")
```

##### La france est donc bien le 2ème pays à avoir organisé le plus de JO après les USA

```python
# 3. Les JO d’hiver sont nés à Chamonix en 1924 ?

# Filtrer les Jeux Olympiques d'hiver
winter_olympics = df_hosts[df_hosts['game_season'] == 'Winter']

# Trier par année pour trouver le premier Jeux d'hiver organisé
first_winter_olympics = winter_olympics.sort_values(by='game_year').iloc[0]

# Construire le message de réponse
location = first_winter_olympics['game_location']
year = first_winter_olympics['game_year']
game_name = first_winter_olympics['game_name']

# Afficher le résultat
print(f"Les premiers JO d'Hiver ont eu lieu à {game_name} en {year}")
```

```python
# 4. JO de Paris, en 1900 : les femmes peuvent participer aux JO ?

# Filtrer les médailles remportées lors des Jeux de Paris 1900
paris_1900_medals = df_medals[df_medals['slug_game'] == 'paris-1900']

# Vérifier la présence de compétitions féminines
women_events = paris_1900_medals[paris_1900_medals['event_gender'] == 'Women']

# Afficher les résultats
if not women_events.empty:
    print(f"Des compétitions féminines ont eu lieu lors des JO de Paris 1900.")
    print(f"Événements: {women_events['event_title'].unique()}")
else:
    print("Aucune compétition féminine n'a été enregistrée lors des JO de Paris 1900.")
```

```python
# 6. De 1924 à 1992, les JO d’hiver et d’été avaient lieu au cours de la même année. Désormais, ils sont organisés selon des cycles distincts avec une alternance de 2 ans ?

# Convertir les années en entiers
df_hosts['game_year'] = df_hosts['game_year'].astype(int)

# Filtrer les données pour les Jeux entre 1924 et 1992
games_1924_1992 = df_hosts[(df_hosts['game_year'] >= 1924) & (df_hosts['game_year'] <= 1992)]

# Vérifier si les Jeux d'hiver et d'été avaient lieu la même année
same_year_games = games_1924_1992.pivot_table(index='game_year', columns='game_season', aggfunc='size', fill_value=0)
same_year_games = same_year_games[(same_year_games['Summer'] > 0) & (same_year_games['Winter'] > 0)]

# Afficher les résultats
if same_year_games.empty:
    print("Entre 1924 et 1992, les JO d'hiver et d'été n'ont pas eu lieu la même année.")
else:
    print(same_year_games)
    print("Entre 1924 et 1992, les JO d'hiver et d'été ont bien eu lieu la même année.")


# Vérifier s'ils sont désormais organisés selon des cycles distincts avec une alternance de 2 ans après 1992
games_after_1992 = df_hosts[df_hosts['game_year'] > 1992]
games_after_1992 = games_after_1992.pivot_table(index='game_year', columns='game_season', aggfunc='size', fill_value=0)

# Afficher les résultats
if games_after_1992.empty:
    print("Après 1992, les JO d'hiver et d'été n'ont pas eu lieu la même année.")
else:
    print(games_after_1992)
    print("Après 1992, les JO d'hiver et d'été sont organisés selon des cycles distincts avec une alternance de 2 ans.")
```

```python
# 7. Tarzan lui-même a participé aux JO. En effet, Johnny Weissmuller, ancien athlète devenu acteur et ayant interprété 12 films de Tarzan, a remporté 5 médailles d’or en natation dans les années 1920 ?

# Filtrer les médailles d'or remportées par Johnny Weissmuller
weissmuller_gold_medals = df_medals[(df_medals['athlete_full_name'] == 'Johnny WEISSMULLER') & (df_medals['medal_type'] == 'GOLD')]

# Afficher le nombre de médailles d'or et les années d'obtention
if not weissmuller_gold_medals.empty:
    print(f"Johnny Weissmuller a remporté {len(weissmuller_gold_medals)} médailles d'or aux JO. Voici les détails :")
    for index, row in weissmuller_gold_medals.iterrows():
        print(f"Médaille d'Or - {row['event_title']}, Sport : {row['discipline_title']}, Jeux : {row['slug_game']}")
else:
    print("Aucune médaille d'or n'a été remportée par Johnny Weissmuller aux JO selon les données disponibles.")

if len(weissmuller_gold_medals) >= 5:
    print("Johnny Weissmuller a bien remporté 5 médailles d'or aux JO.")
else:
    print("Johnny Weissmuller n'a pas remporté 5 médailles d'or aux JO.")
```

```python
# 8. Les JO de Londres de 2012 ont été les 1ers JO durant lesquels tous les pays participants ont envoyé des athlètes de sexe féminin ?

# Filtrer les données pour les Jeux de Londres 2012
london_2012 = df_medals[df_medals['slug_game'] == 'london-2012']

# Vérifier si tous les pays ont envoyé des athlètes féminines
female_participants = london_2012[london_2012['event_gender'] == 'Women']
countries_with_female_athletes = female_participants['country_name'].unique()

# Obtenir la liste des pays participants à Londres 2012
all_countries = london_2012['country_name'].unique()

# Comparer les deux listes pour voir si tous les pays ont envoyé des athlètes féminines
if set(countries_with_female_athletes) == set(all_countries):
    print("Tous les pays participants aux JO de Londres 2012 ont envoyé des athlètes féminines.")
else:
    missing_countries = set(all_countries) - set(countries_with_female_athletes)
    print(f"Les pays suivants n'ont pas envoyé d'athlètes féminines aux JO de Londres 2012 : {missing_countries}")
```

```python
# 10. Les Jeux de 2016, à Rio, marqueront la 1è manifestation des JO en Amérique du Sud ?

 # Convertir les années en entiers pour faciliter les filtrages
df_hosts['game_year'] = df_hosts['game_year'].astype(int)

# Filtrer pour voir si des Jeux ont été organisés en Amérique du Sud avant 2016
south_american_games_before_2016 = df_hosts[(df_hosts['game_location'].str.contains("Brazil|Argentina|Peru|Chile|Ecuador|Colombia|Venezuela", regex=True)) & (df_hosts['game_year'] < 2016)]

# Vérifier si la liste est vide
if south_american_games_before_2016.empty:
    print("Les Jeux de 2016 à Rio marquent la première manifestation des JO en Amérique du Sud.")
else:
    print("Il y a eu des Jeux Olympiques en Amérique du Sud avant 2016. Voici les détails :")
    print(south_american_games_before_2016)
```
