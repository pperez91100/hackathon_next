```python
import xml.etree.ElementTree as ET
import json
import pandas as pd
```

```python
#JSON
json_file_path = './Hackathon_Data/olympic_athletes.json'

# Ouvrir et lire le fichier JSON
with open(json_file_path, 'r', encoding='utf-8') as file:
    json_data = json.load(file)

print(json_data[:3])
```

```python
df_athlete = pd.DataFrame(json_data)
df_athlete
```

```python
#XML
xml_file_path = './Hackathon_Data/olympic_hosts.xml'

# Lire et parser le fichier XML
tree = ET.parse(xml_file_path)
root = tree.getroot()

#convertir les donnÃ©es en une structure de dictionnaire
data = []
for row in root.findall('row'):
    row_data = {subchild.tag: subchild.text for subchild in row}
    data.append(row_data)
```

```python
# xml dans un dataframe
df_hosts = pd.DataFrame(data)
df_hosts
```

```python
# Combien de mÃ©dailles la France a remportÃ© : en tout, en Or, en argent et en Bronze (depuis le dÃ©but des JO) ?

# on filtre sur la France
france_medals = df_medals[df_medals['country_name'] == 'France']

# nombre total de mÃ©dailles
total_medals = france_medals.shape[0]

#nombre de mÃ©dailles d'or, d'argent et de bronze
gold_medals = france_medals[france_medals['medal_type'] == 'GOLD'].shape[0]
silver_medals = france_medals[france_medals['medal_type'] == 'SILVER'].shape[0]
bronze_medals = france_medals[france_medals['medal_type'] == 'BRONZE'].shape[0]

print(f"Total des mÃ©dailles remportÃ©es par la France : {total_medals}")
print(f"Total des mÃ©dailles d'or : {gold_medals}")
print(f"Total des mÃ©dailles d'argent : {silver_medals}")
print(f"Total des mÃ©dailles de bronze : {bronze_medals}")
```

```python
# Lors de quelle JO la France a eu le plus (le moins) de succÃ¨s ?

# filtre sur la France
france_medals = df_medals[df_medals['country_name'] == 'France']

#nombre de mÃ©dailles pour chaque Ã©dition
medals_per_game = france_medals.groupby('slug_game').size().reset_index(name='medal_count')

# le plus de mÃ©dailles
max_medals = medals_per_game[medals_per_game['medal_count'] == medals_per_game['medal_count'].max()]

#le moins de mÃ©dailles
min_medals = medals_per_game[medals_per_game['medal_count'] == medals_per_game['medal_count'].min()]

print("Ã‰dition des JO avec le plus de succÃ¨s pour la France:")
print(max_medals)
print("Ã‰dition des JO avec le moins de succÃ¨s pour la France:")
print(min_medals)
```

```python
# nombre de mÃ©daille obtenues par la France pour chaque JO
medals_per_game_sorted = medals_per_game.sort_values(by='medal_count')
medals_per_game_sorted
```

```python
# Peut-on considÃ©rer que la France est la grande spÃ©cialiste de certaine(s) discipline(s) sportive(s),
# comparativement aux autres nations ? Laquelle / Lesquelles ?

# nombre de mÃ©dailles par discipline
medals_per_discipline = france_medals.groupby('discipline_title').size().reset_index(name='medal_count')

excellent_disciplines = medals_per_discipline[medals_per_discipline['medal_count'] >= 50]

print("Meilleur Disciplines de la France :")
print(excellent_disciplines)
```

```python
# Quelles sont les sports les plus dominants dans les JO au fil des ans ?

#nombre total de mÃ©dailles par sport
medals_per_sport = df_medals.groupby('discipline_title').size().reset_index(name='medal_count')

#trier les sports par le nombre total de mÃ©dailles
most_dominant_sports = medals_per_sport.sort_values(by='medal_count', ascending=False)

print("Sports les plus dominants dans les Jeux Olympiques :")
most_dominant_sports.head(10)
```

```python
# 2. La France est le 2Ã¨ pays qui a organisÃ© le plus de JO aprÃ¨s les USA (8 JO) ?

# Compter le nombre de fois que chaque pays a organisÃ© les JO
hosts_count = df_hosts['game_location'].value_counts()

# Trouver le rang de la France et comparer au nombre total pour les USA
france_rank = hosts_count.rank(method='max', ascending=False)['France']
usa_organized = hosts_count['United States']

# Afficher les rÃ©sultats
print(f"La France est classÃ©e {int(france_rank)}Ã¨me en nombre de JO organisÃ©s.")
print(f"Les Ã‰tats-Unis ont organisÃ© {usa_organized} JO.")
```

```python
# 4. JO de Paris, en 1900 : les femmes peuvent participer aux JO ?

# Filtrer les mÃ©dailles remportÃ©es lors des Jeux de Paris 1900
paris_1900_medals = df_medals[df_medals['slug_game'] == 'paris-1900']

# VÃ©rifier la prÃ©sence de compÃ©titions fÃ©minines
women_events = paris_1900_medals[paris_1900_medals['event_gender'] == 'Women']

# Afficher les rÃ©sultats
if not women_events.empty:
    print(f"Des compÃ©titions fÃ©minines ont eu lieu lors des JO de Paris 1900.")
    print(f"Ã‰vÃ©nements: {women_events['event_title'].unique()}")
else:
    print("Aucune compÃ©tition fÃ©minine n'a Ã©tÃ© enregistrÃ©e lors des JO de Paris 1900.")
```

```python
# 6. De 1924 Ã  1992, les JO dâ€™hiver et dâ€™Ã©tÃ© avaient lieu au cours de la mÃªme annÃ©e. DÃ©sormais, ils sont organisÃ©s selon des cycles distincts avec une alternance de 2 ans ?

# Convertir les annÃ©es en entiers
df_hosts['game_year'] = df_hosts['game_year'].astype(int)

# Filtrer les donnÃ©es pour les Jeux entre 1924 et 1992
games_1924_1992 = df_hosts[(df_hosts['game_year'] >= 1924) & (df_hosts['game_year'] <= 1992)]

# VÃ©rifier si les Jeux d'hiver et d'Ã©tÃ© avaient lieu la mÃªme annÃ©e
same_year_games = games_1924_1992.pivot_table(index='game_year', columns='game_season', aggfunc='size', fill_value=0)
same_year_games = same_year_games[(same_year_games['Summer'] > 0) & (same_year_games['Winter'] > 0)]

# Afficher les rÃ©sultats
if same_year_games.empty:
    print("Entre 1924 et 1992, les JO d'hiver et d'Ã©tÃ© n'ont pas eu lieu la mÃªme annÃ©e.")
else:
    print(same_year_games)
    print("Entre 1924 et 1992, les JO d'hiver et d'Ã©tÃ© ont bien eu lieu la mÃªme annÃ©e.")


# VÃ©rifier s'ils sont dÃ©sormais organisÃ©s selon des cycles distincts avec une alternance de 2 ans aprÃ¨s 1992
games_after_1992 = df_hosts[df_hosts['game_year'] > 1992]
games_after_1992 = games_after_1992.pivot_table(index='game_year', columns='game_season', aggfunc='size', fill_value=0)

# Afficher les rÃ©sultats
if games_after_1992.empty:
    print("AprÃ¨s 1992, les JO d'hiver et d'Ã©tÃ© n'ont pas eu lieu la mÃªme annÃ©e.")
else:
    print(games_after_1992)
    print("AprÃ¨s 1992, les JO d'hiver et d'Ã©tÃ© sont organisÃ©s selon des cycles distincts avec une alternance de 2 ans.")
```

```python
# 7. Tarzan lui-mÃªme a participÃ© aux JO. En effet, Johnny Weissmuller, ancien athlÃ¨te devenu acteur et ayant interprÃ©tÃ© 12 films de Tarzan, a remportÃ© 5 mÃ©dailles dâ€™or en natation dans les annÃ©es 1920 ?

# Filtrer les mÃ©dailles d'or remportÃ©es par Johnny Weissmuller
weissmuller_gold_medals = df_medals[(df_medals['athlete_full_name'] == 'Johnny WEISSMULLER') & (df_medals['medal_type'] == 'GOLD')]

# Afficher le nombre de mÃ©dailles d'or et les annÃ©es d'obtention
if not weissmuller_gold_medals.empty:
    print(f"Johnny Weissmuller a remportÃ© {len(weissmuller_gold_medals)} mÃ©dailles d'or aux JO. Voici les dÃ©tails :")
    for index, row in weissmuller_gold_medals.iterrows():
        print(f"MÃ©daille d'Or - {row['event_title']}, Sport : {row['discipline_title']}, Jeux : {row['slug_game']}")
else:
    print("Aucune mÃ©daille d'or n'a Ã©tÃ© remportÃ©e par Johnny Weissmuller aux JO selon les donnÃ©es disponibles.")

if len(weissmuller_gold_medals) >= 5:
    print("Johnny Weissmuller a bien remportÃ© 5 mÃ©dailles d'or aux JO.")
else:
    print("Johnny Weissmuller n'a pas remportÃ© 5 mÃ©dailles d'or aux JO.")
```

```python
# 8. Les JO de Londres de 2012 ont Ã©tÃ© les 1ers JO durant lesquels tous les pays participants ont envoyÃ© des athlÃ¨tes de sexe fÃ©minin ?

# Filtrer les donnÃ©es pour les Jeux de Londres 2012
london_2012 = df_medals[df_medals['slug_game'] == 'london-2012']

# VÃ©rifier si tous les pays ont envoyÃ© des athlÃ¨tes fÃ©minines
female_participants = london_2012[london_2012['event_gender'] == 'Women']
countries_with_female_athletes = female_participants['country_name'].unique()

# Obtenir la liste des pays participants Ã  Londres 2012
all_countries = london_2012['country_name'].unique()

# Comparer les deux listes pour voir si tous les pays ont envoyÃ© des athlÃ¨tes fÃ©minines
if set(countries_with_female_athletes) == set(all_countries):
    print("Tous les pays participants aux JO de Londres 2012 ont envoyÃ© des athlÃ¨tes fÃ©minines.")
else:
    missing_countries = set(all_countries) - set(countries_with_female_athletes)
    print(f"Les pays suivants n'ont pas envoyÃ© d'athlÃ¨tes fÃ©minines aux JO de Londres 2012 : {missing_countries}")
```

```python
# 10. Les Jeux de 2016, Ã  Rio, marqueront la 1Ã¨ manifestation des JO en AmÃ©rique du Sud ?

 # Convertir les annÃ©es en entiers pour faciliter les filtrages
df_hosts['game_year'] = df_hosts['game_year'].astype(int)

# Filtrer pour voir si des Jeux ont Ã©tÃ© organisÃ©s en AmÃ©rique du Sud avant 2016
south_american_games_before_2016 = df_hosts[(df_hosts['game_location'].str.contains("Brazil|Argentina|Peru|Chile|Ecuador|Colombia|Venezuela", regex=True)) & (df_hosts['game_year'] < 2016)]

# VÃ©rifier si la liste est vide
if south_american_games_before_2016.empty:
    print("Les Jeux de 2016 Ã  Rio marquent la premiÃ¨re manifestation des JO en AmÃ©rique du Sud.")
else:
    print("Il y a eu des Jeux Olympiques en AmÃ©rique du Sud avant 2016. Voici les dÃ©tails :")
    print(south_american_games_before_2016)
```
